```
import pandas as pd
```

```
headers = ['day','chicago','chennai','berlin']
data = """Monday,32,75,41
Tuesday,30,77,43
Wednesday,28,75,45
Thursday,22,82,38
Friday,30,83,30
Saturday,20,81,45
Sunday,25,77,47"""
```

```
dictdata = []
for line in data.splitlines():
    dictdata.append(dict(zip(headers, line.split(','))))
```

```
df = pd.DataFrame(dictdata)
df
```

```
df.melt(id_vars=['day'])
```

```
df.melt(id_vars=['day'], var_name=['city'], value_name="temperature")
```
