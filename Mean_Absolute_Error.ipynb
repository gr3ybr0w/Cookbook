```
fig = plt.figure()
ax = plt.axes()
x = np.linspace(-10, 10, 1000)
ax.plot(x, 1.2 * x + 2)
ax.scatter([2, 5, -4, -7, 8], [-2, 6, -4, 1, 14], marker='o', color="green");
```

```
x = np.array([2, 5, -4, -7, 8])
y = np.array([-2, 6, -4, 1, 14])
yHat = 1.2 * x + 2
```

```
np.divide(np.sum(np.abs(y - yHat)), len(x))
```
